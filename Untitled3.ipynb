# Импорты и устройство

```python
import os
import random
import numpy as np
import matplotlib.pyplot as plt

import torch
import torch.nn as nn
from torch.utils.data import DataLoader, Subset
from torchvision import datasets, transforms, models

from PIL import Image
from tqdm import tqdm
from sklearn.metrics import confusion_matrix
import seaborn as sns  # только для красивой матрицы

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Используемое устройство:", device)
```

# Трансформации и полный датасет

```python
images_root = "/content/CUB_200_2011/images"

# Трансформации: train с аугментацией, valid/test только resize+normalize
transform_train = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.RandomHorizontalFlip(p=0.5),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406],
                         std=[0.229, 0.224, 0.225]),
])

transform_eval = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406],
                         std=[0.229, 0.224, 0.225]),
])

# Полный датасет: отдельно для train (с аугментацией) и eval (без)
dataset_train_full = datasets.ImageFolder(root=images_root, transform=transform_train)
dataset_eval_full  = datasets.ImageFolder(root=images_root, transform=transform_eval)

class_names = dataset_train_full.classes
num_classes = len(class_names)

print("Классов:", num_classes)
print("Примеры классов:", class_names[:5])
print("Всего изображений:", len(dataset_train_full))
```

# Быстрый сплит на train/valid/test (subset + рандом)

```python
from torch.utils.data import Subset

# Фиксируем рандом для воспроизводимости
random_seed = 42
np.random.seed(random_seed)
random.seed(random_seed)
torch.manual_seed(random_seed)

n_total = len(dataset_train_full)
indices = np.random.permutation(n_total)

# Ограничиваем реальный размер выборок, чтобы обучение было быстрым
max_train = 2000   # можно увеличить, если Colab тянет
max_valid = 500
max_test  = 500

train_idx_small = indices[:max_train]
valid_idx_small = indices[max_train:max_train + max_valid]
test_idx_small  = indices[max_train + max_valid : max_train + max_valid + max_test]

print("Train small:", len(train_idx_small))
print("Valid small:", len(valid_idx_small))
print("Test  small:", len(test_idx_small))

train_dataset = Subset(dataset_train_full, train_idx_small)
valid_dataset = Subset(dataset_eval_full,  valid_idx_small)
test_dataset  = Subset(dataset_eval_full,  test_idx_small)

batch_size = 32
num_workers = 2  # можно 0, если будут проблемы

train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True,  num_workers=num_workers)
valid_loader = DataLoader(valid_dataset, batch_size=batch_size, shuffle=False, num_workers=num_workers)
test_loader  = DataLoader(test_dataset,  batch_size=batch_size, shuffle=False, num_workers=num_workers)
```

# Визуализация батча

```python
def show_batch(images, labels, class_names, n=6):
    images = images[:n]
    labels = labels[:n]
    images_np = images.permute(0, 2, 3, 1).cpu().numpy()

    mean = np.array([0.485, 0.456, 0.406])
    std  = np.array([0.229, 0.224, 0.225])

    plt.figure(figsize=(15, 4))
    for i in range(len(images_np)):
        img = images_np[i] * std + mean
        img = np.clip(img, 0, 1)
        plt.subplot(1, n, i+1)
        plt.imshow(img)
        plt.axis("off")
        plt.title(class_names[labels[i].item()])
    plt.show()

images, labels = next(iter(train_loader))
show_batch(images, labels, class_names, n=6)
```

# Модель ResNet18 (предобученная, только fc обучается)

```python
from torchvision.models import ResNet18_Weights

weights = ResNet18_Weights.IMAGENET1K_V1
model = models.resnet18(weights=weights)

# Замораживаем все слои, кроме последнего
for param in model.parameters():
    param.requires_grad = False

# Меняем финальный слой под 200 классов
in_features = model.fc.in_features
model.fc = nn.Linear(in_features, num_classes)

model = model.to(device)

criterion = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(model.fc.parameters(), lr=1e-3)

model
```

# train/eval с прогресс-баром и top-1 / top-5

```python
def train_one_epoch(model, loader, optimizer, criterion, device, epoch_idx, num_epochs):
    model.train()
    running_loss = 0.0
    running_corrects = 0
    total = 0

    progress_bar = tqdm(loader, desc=f"Train {epoch_idx+1}/{num_epochs}", leave=False)

    for images, labels in progress_bar:
        images = images.to(device)
        labels = labels.to(device)

        optimizer.zero_grad()
        outputs = model(images)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()

        # top-1 accuracy
        _, preds = torch.max(outputs, 1)
        running_loss += loss.item() * images.size(0)
        running_corrects += torch.sum(preds == labels).item()
        total += labels.size(0)

        current_loss = running_loss / total
        current_acc = running_corrects / total
        progress_bar.set_postfix(loss=f"{current_loss:.4f}", acc=f"{current_acc*100:.1f}%")

    epoch_loss = running_loss / total
    epoch_acc = running_corrects / total
    return epoch_loss, epoch_acc


def evaluate(model, loader, criterion, device, mode="Val", compute_top5=False):
    model.eval()
    running_loss = 0.0
    running_corrects_top1 = 0
    running_corrects_top5 = 0
    total = 0

    all_labels = []
    all_preds = []

    with torch.no_grad():
        progress_bar = tqdm(loader, desc=f"{mode}", leave=False)
        for images, labels in progress_bar:
            images = images.to(device)
            labels = labels.to(device)

            outputs = model(images)
            loss = criterion(outputs, labels)

            # top-1
            _, preds_top1 = torch.max(outputs, 1)

            running_loss += loss.item() * images.size(0)
            running_corrects_top1 += torch.sum(preds_top1 == labels).item()

            if compute_top5:
                # top-5
                _, top5 = outputs.topk(5, dim=1)
                correct_top5 = top5.eq(labels.view(-1, 1)).any(dim=1)
                running_corrects_top5 += correct_top5.sum().item()

            total += labels.size(0)

            current_loss = running_loss / total
            current_acc1 = running_corrects_top1 / total
            if compute_top5:
                current_acc5 = running_corrects_top5 / total
                progress_bar.set_postfix(loss=f"{current_loss:.4f}",
                                         top1=f"{current_acc1*100:.1f}%",
                                         top5=f"{current_acc5*100:.1f}%")
            else:
                progress_bar.set_postfix(loss=f"{current_loss:.4f}",
                                         top1=f"{current_acc1*100:.1f}%")

            # Для confusion matrix
            all_labels.extend(labels.cpu().numpy())
            all_preds.extend(preds_top1.cpu().numpy())

    epoch_loss = running_loss / total
    top1_acc = running_corrects_top1 / total
    top5_acc = running_corrects_top5 / total if compute_top5 else None

    return epoch_loss, top1_acc, top5_acc, np.array(all_labels), np.array(all_preds)
```

# Обучение модели (с логом)

```python
num_epochs = 3  # можно уменьшить до 2, если Colab слабый

history = {
    "train_loss": [],
    "train_acc": [],
    "valid_loss": [],
    "valid_top1": [],
    "valid_top5": [],
}

for epoch in range(num_epochs):
    print(f"\n===== Epoch {epoch+1}/{num_epochs} =====")

    train_loss, train_acc = train_one_epoch(
        model, train_loader, optimizer, criterion, device,
        epoch_idx=epoch, num_epochs=num_epochs
    )

    val_loss, val_top1, val_top5, _, _ = evaluate(
        model, valid_loader, criterion, device,
        mode="Val", compute_top5=True
    )

    history["train_loss"].append(train_loss)
    history["train_acc"].append(train_acc)
    history["valid_loss"].append(val_loss)
    history["valid_top1"].append(val_top1)
    history["valid_top5"].append(val_top5)

    print(f"Train  loss: {train_loss:.4f} | acc: {train_acc*100:.2f}%")
    print(f"Valid  loss: {val_loss:.4f} | top-1: {val_top1*100:.2f}% | top-5: {val_top5*100:.2f}%")
```

# Графики метрик

```python
epochs_range = range(1, num_epochs + 1)

plt.figure(figsize=(12,4))

plt.subplot(1,2,1)
plt.plot(epochs_range, history["train_loss"], label="train_loss")
plt.plot(epochs_range, history["valid_loss"], label="valid_loss")
plt.xlabel("Эпоха")
plt.ylabel("Loss")
plt.title("Loss (train vs valid)")
plt.legend()

plt.subplot(1,2,2)
plt.plot(epochs_range, np.array(history["train_acc"])*100, label="train_top1")
plt.plot(epochs_range, np.array(history["valid_top1"])*100, label="valid_top1")
plt.plot(epochs_range, np.array(history["valid_top5"])*100, label="valid_top5")
plt.xlabel("Эпоха")
plt.ylabel("Accuracy (%)")
plt.title("Accuracy (top-1 / top-5)")
plt.legend()

plt.show()
```

# Оценка на тесте с top-5 и сбор предсказаний

```python
test_loss, test_top1, test_top5, y_true, y_pred = evaluate(
    model, test_loader, criterion, device,
    mode="Test", compute_top5=True
)

print(f"\n==== Итоги на тесте ====")
print(f"Test loss:   {test_loss:.4f}")
print(f"Test top-1:  {test_top1*100:.2f}%")
print(f"Test top-5:  {test_top5*100:.2f}%")
```

# Confusion matrix (200 классов) + топ-20 классов

```python
# Полная confusion matrix (200x200)
cm = confusion_matrix(y_true, y_pred, labels=list(range(num_classes)))
cm_norm = cm.astype("float") / (cm.sum(axis=1, keepdims=True) + 1e-12)

plt.figure(figsize=(10, 8))
plt.imshow(cm_norm, interpolation="nearest", cmap="Blues", aspect="auto")
plt.title("Confusion Matrix (нормированная, 200 классов)")
plt.colorbar()
plt.xlabel("Предсказанный класс")
plt.ylabel("Истинный класс")
plt.tight_layout()
plt.show()

# Топ-20 самых частых классов (чтобы была читаемая красивая матрица)
class_counts = cm.sum(axis=1)
top_k = 20
top_classes_idx = np.argsort(class_counts)[-top_k:]

cm_top = cm_norm[np.ix_(top_classes_idx, top_classes_idx)]
labels_top = [class_names[i] for i in top_classes_idx]

plt.figure(figsize=(12, 10))
sns.heatmap(cm_top, annot=False, cmap="Blues",
            xticklabels=labels_top, yticklabels=labels_top)
plt.xticks(rotation=90)
plt.yticks(rotation=0)
plt.title(f"Confusion Matrix — топ-{top_k} самых частых классов")
plt.tight_layout()
plt.show()
```

# Предсказание по одной картинке

```python
inference_transform = transform_eval  # тот же, что для eval

def predict_image(path, model, class_names, device):
    model.eval()
    img = Image.open(path).convert("RGB")
    x = inference_transform(img).unsqueeze(0).to(device)

    with torch.no_grad():
        outputs = model(x)
        probs = torch.softmax(outputs, dim=1)
        top5_prob, top5_idx = probs.topk(5, dim=1)

    top5_prob = top5_prob.cpu().numpy().flatten()
    top5_idx = top5_idx.cpu().numpy().flatten()

    plt.imshow(img)
    plt.axis("off")
    plt.show()

    print("Top-5 предсказаний:")
    for p, idx in zip(top5_prob, top5_idx):
        print(f"{class_names[idx]} — {p*100:.2f}%")

# Берём случайное изображение из тестовой выборки
all_test_paths = []
for idx in test_idx_small:
    path, _ = dataset_eval_full.samples[idx]
    all_test_paths.append(path)

sample_path = random.choice(all_test_paths)
print("Пример файла:", sample_path)
predict_image(sample_path, model, class_names, device)
```
